```python
import pandas as pd

# Read the JSON file
data_orders = pd.read_json('orders.json')
data_countries = pd.read_json('countries.json')
data_products = pd.read_json('products.json')
data_customers = pd.read_json('customers.json')
data_sales = pd.read_json('sales.json')

print(data_countries)

# Convert JSON to Excel
data_orders.to_excel('output_orders.xlsx', index=False)
data_countries.to_excel('output_countries.xlsx', index=False)
data_products.to_excel('output_products.xlsx', index=False)
data_customers.to_excel('output_customers.xlsx', index=False)
data_sales.to_excel('output_sales.xlsx', index=False)

```
